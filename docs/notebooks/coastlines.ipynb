from mapgpye import GeoMap


```python
from ecmwf.geomaps import GeoMap
```

```python
geomap = GeoMap()
```

## Add a coastlines object 

```python
geomap.coastlines()
```

##  Call the show method to render the object
The coastlines will be rendered using all the  defaut values on the default geographical area/projection ie over the globe on cylindrical projection : EPSG:4326

```python
geomap.show()
```

## Tailor your plot
 -  Change the visual attributes of the coastlines 
 -  Add gridlines, borders, rivers 

```python
geomap = GeoMap()

geomap.coastlines(land_colour="cream")
geomap.gridlines(line_colour='grey', line_style="dash")

geomap.show()
```
