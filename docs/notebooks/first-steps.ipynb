````python
from ecmwf.geomaps import GeoMap
````

## Create your first map

````python
geomap = GeoMap()
````

The coastlines will be rendered using all the  defaut values on the default geographical area/projection ie over the globe on cylindrical projection : EPSG:4326

````python
geomap.show()
````

## Tailor your plot
 -  Change the visual attributes of the [coastlines](https://magpye-test.readthedocs.io/en/latest/reference/api.html#ecmwf.geomaps.GeoMap.coastlines) 
 -  Add [gridlines](https://magpye-test.readthedocs.io/en/latest/reference/api.html#ecmwf.geomaps.GeoMap.gridlines), borders, [rivers](https://magpye-test.readthedocs.io/en/latest/reference/api.html#ecmwf.geomaps.GeoMap.rivers)

````python
geomap = GeoMap()

geomap.coastlines(land_colour="grey")
geomap.gridlines(line_colour='grey', 
                 line_style="dash")

geomap.show()
````

## Change the geographical area

Magpye offers a set of predefined areas.
You can select one when creating the map.

````python
geomap = GeoMap(area_name = "europe")

geomap.coastlines(land_colour="grey")
geomap.gridlines(line_colour='grey', 
                 line_style="dash")

geomap.show()
````

## Add a geographical field

Your data is in a file called mslp.grib --> mean sea level pressure coded in grib.  
This default visualisation is a set of blue isolines

````python
geomap = GeoMap(area_name = "europe")

geomap.coastlines(land_colour="grey")
geomap.contour_lines("mslp.grib")
geomap.gridlines(line_colour='grey', 
                 line_style="dash")

geomap.show()
````

## Tweak the plot

You can now change the visualisation settings of the [contour_lines](https://magpye-test.readthedocs.io/en/latest/reference/api.html#ecmwf.geomaps.GeoMap.contour_lines) method.
In this example :  
- We will increase the sresolution of the coastlines
- Change the colour of some isolines and highlight some in purple 
- Plot a line every 2 millibars (starting at 0) with highlights every 4 lines
- Add a title 

````python

geomap = GeoMap(area_name="europe")

geomap.coastlines(resolution="high", land_colour="grey")
geomap.contour_lines(
    source="mslp.grib",
    line_colour="red", highlight_colour="purple",
    interval=2, interval_reference=0, highlight_frequency=4,
)

geomap.coastlines(resolution="high")
geomap.gridlines(line_style="dash", line_colour="#555555")

geomap.title("Mean sea level pressure at 00 UTC on Tuesday 30/10/2012")

geomap.show()
````
