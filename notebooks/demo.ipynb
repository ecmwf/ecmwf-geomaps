```python
from magpye import GeoMap

HRES_MSL = "../tests/data/hres_msl.grib"
```

## Global map of sea level pressure

```python
# Instantiate a new map
geomap = GeoMap()

# Add some background land polygons and coastlines
geomap.coastlines(land_colour="cream")

# Plot simple, default (blue) line contours on the map
geomap.contour_lines(HRES_MSL)

# Add foreground coastlines and gridlines
geomap.coastlines()
geomap.gridlines()

# Add an auto-generated title
geomap.title()

# Show the finished map inline in the notebook
geomap.show()
```

### Tweaking the plot

```python
# Instantiate a new map, this time using the default Europe view
geomap = GeoMap(area_name="europe")

# Add some higher-resolution background land polygons and coastlines
geomap.coastlines(resolution="high", land_colour="cream")

# Plot red line contours on the map, with purple highlights
# Plot a line every 2 millibars (starting at 0) with highlights every 4 lines
geomap.contour_lines(
    source=HRES_MSL,
    line_colour="red",
    highlight_colour="purple",
    interval=200,
    interval_reference=0,
    highlight_frequency=4,
)

# Add foreground coastlines and gridlines
geomap.coastlines(resolution="high")
geomap.gridlines(line_style="dash", line_colour="#555555")

# Add a custom title
geomap.title("Mean sea level pressure at 00 UTC on Tuesday 30/10/2012")

# Show the finished map inline in the notebook
geomap.show()
```

## Applying styles

### Pre-defined styles

```python
import ecmwf.data as ecdata

msl = ecdata.read(HRES_MSL)
```

```python
geomap = GeoMap(area_name="europe")

# contour methods can take a 'style' argument, which automatically
# populates certain arguments with pre-defined values for convenience
geomap.contour_shaded(msl / 100, style="demo")
geomap.coastlines(line_colour="red", resolution="high")
geomap.legend(title="MSL shaded contours")
geomap.title()
geomap.gridlines()
geomap.footer(
    "Hello world", datestamp=True, font_size=0.5, font_colour="black", logo="ecmwf"
)

geomap.show()
```
