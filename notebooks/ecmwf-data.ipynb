```python
from ecmwf.geomaps import GeoMap

HRES_MSL = "../tests/data/hres_msl.grib"
```

```python
import ecmwf.data as ecdata
```

## Reading data into ecmwf-data before plotting

```python

geomap = GeoMap()
geomap.coastlines(land_colour="cream")

# Load some MSL data into ecmwf-data and plot it as contour lines
fs = ecdata.read(HRES_MSL)
geomap.contour_lines(fs)

geomap.coastlines()
geomap.gridlines()
geomap.title()

geomap.show()
```

### Convert to Pa before plotting

```python
geomap = GeoMap()
geomap.coastlines(land_colour="cream")

# Load some MSL data into ecmwf-data
fs = ecdata.read(HRES_MSL)

# Convert to Pa (multiply by 100)
fs *= 100

# Plot contour lines in Pa
geomap.contour_lines(fs)

geomap.coastlines()
geomap.gridlines()
geomap.title()

geomap.show()
```
