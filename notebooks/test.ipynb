```python
from ecmwf.geomaps import GeoMap
```

## Global map of seal level pressure

```python
fig = GeoMap(projection='mollweide')
fig.coastlines(resolution='high', land_colour='cream')
fig.data('msl.grib', colour='red', highlight_colour='purple')
fig.coastlines(resolution='high')
fig.grid()
fig.title('ECMWF Geomap')
fig.show()
```

## Applying method-specific presets

```python
fig = GeoMap(area_name='europe')
fig.coastlines(preset='default-background')
fig.coastlines(preset='default')
fig.grid(preset='default')
fig.show()
```

## Applying global presets

```python
fig = GeoMap(area_name='europe', preset='background-foreground')
fig.show()
```
