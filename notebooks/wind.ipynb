```python
from ecmwf.geomaps import GeoMap

UV_RH = "../tests/data/uv_rh.grib"
```

## Europe map of wind

```python
geomap = GeoMap(area_name="europe", preset="background-foreground")
geomap.wind(UV_RH, wind_style="flags", colour="red", density=2)
geomap.show()
```

```python
geomap = GeoMap(area_name="europe", preset="background-foreground")
geomap.wind(
    UV_RH, wind_style="arrows", arrow_head="angle-45", colour="purple", density=2
)
geomap.show()
```

```python
geomap = GeoMap(area_name="europe", preset="background-foreground")
geomap.wind_shaded(
    UV_RH, wind_style="arrows", arrow_head="triangle-15", bins=list(range(0, 100, 10))
)
geomap.show()
```
