```python
from ecmwf.geomaps import GeoMap
import ecmwf.data as ecdata

UV_RH = "../tests/data/uv_rh.grib"

fs = ecdata.read(UV_RH)
u = fs.select(shortName="u", level=1000)
v = fs.select(shortName="v", level=1000)
```

## Europe map of wind

```python
geomap = GeoMap(area_name="europe", preset="background-foreground")
geomap.arrows(u=u, v=v, wind_style="flags", colour="red", density=2)
geomap.show()
```

```python
geomap = GeoMap(area_name="europe", preset="background-foreground")
geomap.arrows(u=u, v=v, shaded=True, wind_style="arrows", arrow_head="triangle-15")
geomap.show()
```
